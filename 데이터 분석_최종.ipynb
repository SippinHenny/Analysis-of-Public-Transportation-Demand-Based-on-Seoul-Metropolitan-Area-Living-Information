## 3.이변량 분석

* 자료의 종류에 맞게 X --> Y 에 대해서 그래프(시각화)와 가설검정(수치화)를 수행하고 결과를 평가합니다.

* 가설검정시 다음의 항목을 참조하여 수행합니다.
    * 유의수준 : 5%
    * 숫자 --> 숫자 : 상관분석
    * 범주 --> 범주 : 카이제곱검정
    * 범주 --> 숫자 : t검정, 분산분석
    * 숫자 --> 범주 : 로지스틱 회귀모형을 통해, 회귀계수의 P.value로 검정을 수행합니다.
    
상관계수 (우리가 배운 기준)
- 0.5 ~ 1 : 강한
- 0.2 ~ 0.5 : 중간
- 0.1 ~ 0.2 : 약간
- 0 ~ 0.1 : 없음
- 음수 똑같음 

p-value 기준을 0.05로 가져죠? 5%
p-value가 0.05보다 크다 == 귀무가설을 기각 하지 못한다 == 대립가설이 틀렸다(우리가 원하는 바는 대립가설)

- 귀무가설을 기각하려면 우리가 싫어하는 귀무가설을 입증할 증거가 나올 확률이 5% 이하여야 함
- p-value가 그 귀무가설을 입증할 증거가 나올 확률을 의미함


```python
# 자치구별 정류장 수와 노선 수를 내림차순으로 정렬
sorted_df = df_seoul_bus_station.sort_values(by=['정류장수'], ascending=False)

# 시각화
plt.figure(figsize=(14, 6))

# 정류장 수
plt.subplot(1, 2, 1)
sns.barplot(data=sorted_df, x='자치구', y='정류장수', palette='Blues')
plt.title('자치구별 정류장 수')
plt.xlabel('자치구')
plt.ylabel('정류장 수')
plt.xticks(rotation=45)

sorted_df = df_seoul_bus_station.sort_values(by=['노선수'], ascending=False)
# 노선 수
plt.subplot(1, 2, 2)
sns.barplot(data=sorted_df, x='자치구', y='노선수', palette='Oranges')
plt.title('자치구별 노선 수')
plt.xlabel('자치구')
plt.ylabel('노선 수')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

### 1) 가설 1 정류장 수와 승차 총 승객수
    * H0 정류장 수와 승차 총 승객 수는 연관이 없다.
    * H1 정류장 수와 승차 총 승객 수는 연관이 있다.

```python
import scipy.stats as spst
spst.pearsonr(df['정류장수'],df['승차총승객수'])
# p-value가 0.05보다 작다 == 대립가설이 맞다. 노선 수와 총 이동 시간은 연관이 있다.
```

### 2)가설 2 노선 수와 승차 총 승객 수
    * H0 노선 수와 승차 총 승객수는 연관이 없다.
    * H1 노선 수와 승차 총 승객수는 연관이 있다.

```python
spst.pearsonr(df['노선수'],df['승차총승객수'])
# p-value가 0.05보다 작다 == 대립가설이 맞다. 정류장 수와 승차 총 승객수는 연관이 있다.
```

### 3)가설 3 정류장 수와 거주 인구 수
    * H0 정류장 수와 거주 인구 수는 연관이 없다.
    * H1 정류장 수와 거주 인구 수는 연관이 있다.

```python
spst.pearsonr(df['정류장수'],df['합계'])
#p-value가 0.05보다 작다 == 대립가설이 맞다. 정류장 수와 승차 총 승객수 연관이 있다.
```

### 4)가설 4 노선 수와 택시 운송업 종사자 수

```python
df['승노 비율'] = df['승차총승객수']/ df['노선수']

#이 비율은 각 노선이 얼마나 많은 승객을 처리하는 지를 보여준다.
#즉 이 비율이 높을수록 노선이 부족하다는 뜻이고 , 해당 지역의 수요가 높다고 볼 수 있다.
```

```python
df_sorted = df.sort_values(by='승노 비율', ascending=False)
df_sorted[['자치구','승노 비율']].head(5)
```

```python
# 노선 수와 정류장 수를 각각 정규화
df['노선수_정규화'] = (df['노선수'] - df['노선수'].min()) / (df['노선수'].max() - df['노선수'].min())
df['정류장수_정규화'] = (df['정류장수'] - df['정류장수'].min()) / (df['정류장수'].max() - df['정류장수'].min())

# 가중 평균을 계산하여 새로운 열 추가
df['노선_정류장_평균'] = (df['노선수_정규화'] + df['정류장수_정규화']) / 2
```

```python
# 승차총승객수를 노선_정류장_평균으로 나눈 열 추가
df['승차승객_노선_정류장_비율'] = df['승차총승객수'] / df['노선_정류장_평균']

# 새로운 열을 기준으로 데이터를 오름차순 정렬
df_sorted = df.sort_values(by='승차승객_노선_정류장_비율', ascending=False)
```
